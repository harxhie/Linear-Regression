```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
print(df["year"].shape)
print(df["per capita income (US$)"].shape)
```

```python
df = df.dropna()  # This removes any rows with missing values in any column

x = df[["year"]]
y = df["per capita income (US$)"]

plt.scatter(x, y)
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Income Over the Years")
plt.show()
```

```python
print(x)
print(y)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[2025]])
```
